### QUES - **17. What is causation? Explain difference between correlation and causation with an example.**  
**ANS -:**  
- **Causation:** Indicates that one event is the result of the occurrence of another.  
- **Difference:**  
  - Correlation: Two variables are related but don't necessarily affect each other.  
  - Causation: A change in one variable directly causes a change in another.  
**Example:**  
- Correlation: Ice cream sales and drowning rates are correlated.  
- Causation: Ice cream sales do not cause drowning; both are related to summer heat.
