**QUES-1:**  
**What is a parameter?**  
**ANS:**  
A parameter is a variable within a machine learning model that is learned from the training data. It represents the relationship between inputs and outputs. For example, in linear regression, parameters include the coefficients that determine the strength of the relationship between features and the target. These parameters are adjusted during the model's training to minimize the error between the predicted and actual outputs.

---

**QUES-2:**  
**What is correlation?**  
**ANS:**  
Correlation is a statistical measure that indicates the strength and direction of the relationship between two variables. The value ranges from -1 to 1:
- **1** means a perfect positive correlation (both variables increase together).
- **-1** means a perfect negative correlation (one variable increases while the other decreases).
- **0** means no correlation.

Correlation helps identify the degree of linear relationship between variables.

---

**QUES-3:**  
**What does negative correlation mean?**  
**ANS:**  
Negative correlation means that as one variable increases, the other decreases, and vice versa. For example, there may be a negative correlation between the number of hours spent watching TV and academic performance, where more TV watching correlates with lower performance.

**Define Machine Learning. What are the main components in Machine Learning?**  
**ANS:**  
Machine learning is a branch of AI where algorithms learn from data and make predictions or decisions without explicit programming. Key components:
1. **Data**: Input used for training the model.
2. **Algorithm**: A method or model to learn from the data.

**QUES-6:**  
**What are continuous and categorical variables?**  
**ANS:**  
1. **Continuous Variables**: These are numerical variables that can take any value within a range (e.g., height, temperature, weight).
2. **Categorical Variables**: These represent distinct categories or groups. Categorical variables can be:
   - **Nominal**: No inherent order (e.g., colors, gender).
   - **Ordinal**: Have a specific order (e.g., education level, rankings).

**QUES-7:**  
**How do we handle categorical variables in Machine Learning? What are the common techniques?**  
**ANS:**  
Categorical variables are often handled using encoding techniques:
1. **One-Hot Encoding**: Converts each category into a binary vector.
2. **Label Encoding**: Assigns a unique integer to each category.
3. **Ordinal Encoding**: Used for ordinal variables, assigns numerical values based on the order.

These techniques help transform categorical data into numerical format, which machine learning models can process.

**QUES-17:**  
**What is causation? Explain difference between correlation and causation with an example.**  
**ANS:**  
Causation implies that one variable directly affects another. In contrast, correlation simply shows that two variables move together but doesn't imply one causes the other. For example, there may be a correlation between ice cream sales and drowning incidents in summer, but the cause is not that ice cream sales increase drowning—it’s the hot weather driving both.

**QUES-18:**  
**What is an Optimizer? What are different types of optimizers? Explain each with an example.**  
**ANS:**  
An optimizer is an algorithm used to minimize the loss function during training. Common optimizers:
1. **Gradient Descent**: Iteratively adjusts parameters in the direction of the steepest decrease in the loss.
2. **Stochastic Gradient Descent (SGD)**: A variant of gradient descent where updates are made after evaluating each data point.
3. **Adam**: Combines the advantages of both AdaGrad and RMSProp for adaptive learning rates.
4. **RMSProp**: Helps optimize neural networks by adjusting the learning rate dynamically.

**ANS:**  
Feature scaling transforms features into a similar range, ensuring no single feature dominates others. Common methods include normalization (scaling between 0 and 1) and standardization (mean = 0, variance = 1). This helps improve the performance of models, especially those sensitive to feature magnitude, like k-NN or gradient descent.
